```python
from numpy import vectorize
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')

# create a simple dataset
data = {
    'email_text' : [
        "free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005",
        " WINNER! click here to claim your price",
        "you have been awarded a $1000 bonuz price"
    ],
    'label' : ['spam', 'ham', 'spam']
}

df = pd.DataFrame(data)

# convert labels to numerical format
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
X = df['email_text']
y = df['label']

# split data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(
    df['email_text'], df['label'], test_size=0.25, random_state=42
)

# convert text into a matrix
vectorizer = CountVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


# initialise and train the RAndom forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_vec, y_train)

#make predictions
y_pred = rf_classifier.predict(X_test_vec)

#test the email
def predict_email(email_text, model, vectorizer):
    email_vec = vectorizer.transform([email_text])
    prediction = model.predict(email_vec)
    return "spam" if prediction[0] == 1 else "ham"

new_email = "hi good evening. meeting will be at 3pm"
print(f"\n new email prediction: {predict_email(new_email, rf_classifier, vectorizer)}")

new_email2 = "can we meet for coffee"
print(f"\n new email prediction: {predict_email(new_email2, rf_classifier, vectorizer)}")
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# download the datasets
from sklearn.datasets import load_diabetes
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# load /read  dataset from url
df = pd.read_csv(url, names=column_names)

# separate features and target
X = df.drop('class', axis=1)
y = df['class']

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# create and train the random forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# make predictions
y_pred = rf.predict(X_test)


# evaluate the model
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred))
```
